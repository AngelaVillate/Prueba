```python
# PASO 1: IMPORTAR LAS LIBRER√çAS NECESARIAS
# =============================================================================
print("\nüîß PASO 1: Preparando nuestras herramientas de trabajo")
print("Las librer√≠as son como cajas de herramientas especializadas:")

import pandas as pd          # Para manejar datos como tablas
import numpy as np           # Para operaciones matem√°ticas
import matplotlib.pyplot as plt  # Para crear gr√°ficos
import seaborn as sns        # Para gr√°ficos m√°s bonitos

print("‚úÖ Librer√≠as importadas correctamente")
print("- pandas: maneja datos como Excel")
print("- numpy: hace c√°lculos matem√°ticos")
print("- matplotlib y seaborn: crean gr√°ficos")
```

```python
# PASO 2: CARGAR LOS DATOS
# =============================================================================
print("\nüìÇ PASO 2: Cargando nuestro dataset")
print("Vamos a abrir el archivo CSV (como una hoja de Excel)")

# Cargar el dataset
ruta_archivo = '/content/Social Media Engagement Dataset.csv'
df = pd.read_csv(ruta_archivo, sep=';', engine='python')

print(f"‚úÖ Archivo cargado exitosamente")
print(f"Nuestros datos tienen {df.shape[0]} filas y {df.shape[1]} columnas")
print("Cada fila representa una publicaci√≥n en redes sociales")
print("Cada columna representa una caracter√≠stica de esa publicaci√≥n")

# Mostramos las primeras 5 filas para tener una idea de los datos
print("\nüëÄ Vistazo a las primeras 5 filas:")
display(df.head())
```

```python
# PASO 3: PRIMERA MIRADA A LOS DATOS
# =============================================================================
print("\nüëÄ PASO 3: Echemos un vistazo a nuestros datos")
print("Como cuando abres un libro, primero ves el √≠ndice:")

print("\nüî∏ Las primeras 5 filas de nuestro dataset:")
print(df.head())

print("\nüî∏ Las columnas (variables) que tenemos:")
print(df.columns.tolist())

print("\nüî∏ Informaci√≥n general sobre nuestros datos:")
print(df.info())
```

```python
# PASO 4: TIPOS DE VARIABLES - CONCEPTO FUNDAMENTAL
# =============================================================================
print("\nüìö PASO 4: Entendiendo los tipos de variables")
print("En estad√≠stica, las variables se clasifican en dos grandes grupos:")
print()
print("üîπ VARIABLES CUALITATIVAS (categ√≥ricas):")
print("   - Describen caracter√≠sticas o categor√≠as")
print("   - Ejemplos: tipo de contenido, plataforma, g√©nero")
print()
print("üîπ VARIABLES CUANTITATIVAS (num√©ricas):")
print("   - Representan cantidades que se pueden medir")
print("   - Ejemplos: n√∫mero de likes, edad, tiempo")

# Identificar tipos autom√°ticamente
variables_numericas = df.select_dtypes(include=[np.number]).columns.tolist()
variables_categoricas = df.select_dtypes(include=['object']).columns.tolist()

print(f"\nüî¢ Variables CUANTITATIVAS en nuestro dataset: {variables_numericas}")
print(f"üìù Variables CUALITATIVAS en nuestro dataset: {variables_categoricas}")
```

```python
# Lista de columnas para analizar
columnas_a_analizar = ['day_of_week', 'platform', 'location']

# Iterar a trav√©s de las columnas y calcular frecuencias
for columna in columnas_a_analizar:
    print(f"An√°lisis de Frecuencias para: {columna}")

    # Frecuencias absolutas
    frecuencias_abs = df[columna].value_counts()
    print("\nFrecuencias Absolutas:")
    print(frecuencias_abs)

    # Frecuencias relativas
    frecuencias_rel = df[columna].value_counts(normalize=True) * 100
    print("\nFrecuencias Relativas (%):")
    # Formatear la salida para que se muestre con el signo de porcentaje
    print(frecuencias_rel.map('{:.2f}%'.format))

    print("-" * 50)

    # Gr√°fico de barras
    plt.figure(figsize=(12, 8))
    sns.countplot(y=columna, data=df, order=df[columna].value_counts().index, palette='viridis')
    plt.title(f'Frecuencia de Publicaciones por {columna}')
    plt.xlabel('Cantidad de Publicaciones')
    plt.ylabel(columna)
    plt.show()
```

```python
# Gr√°fico de barras de frecuencias relativas para 'day_of_week'
plt.figure(figsize=(12, 8))
rel_freq_df_day = df['day_of_week'].value_counts(normalize=True).mul(100).rename('percent').reset_index()
sns.barplot(y='day_of_week', x='percent', data=rel_freq_df_day, palette='viridis')
plt.title('Frecuencia Relativa de Publicaciones por D√≠a de la Semana')
plt.xlabel('Porcentaje de Publicaciones (%)')
plt.ylabel('D√≠a de la Semana')
plt.show()

# Gr√°fico de barras de frecuencias relativas para 'platform'
plt.figure(figsize=(12, 8))
rel_freq_df_platform = df['platform'].value_counts(normalize=True).mul(100).rename('percent').reset_index()
sns.barplot(y='platform', x='percent', data=rel_freq_df_platform, palette='viridis')
plt.title('Frecuencia Relativa de Publicaciones por Plataforma')
plt.xlabel('Porcentaje de Publicaciones (%)')
plt.ylabel('Plataforma')
plt.show()

# Gr√°fico de barras de frecuencias relativas para 'location'
plt.figure(figsize=(12, 10))
rel_freq_df_location = df['location'].value_counts(normalize=True).mul(100).rename('percent').reset_index()
sns.barplot(y='location', x='percent', data=rel_freq_df_location, palette='viridis')
plt.title('Frecuencia Relativa de Publicaciones por Ubicaci√≥n')
plt.xlabel('Porcentaje de Publicaciones (%)')
plt.ylabel('Ubicaci√≥n')
plt.show()
```

```python
# PASO 6: AN√ÅLISIS BIVARIADO - Categ√≥rica vs. Num√©rica
# =============================================================================
print("\nüìä PASO 6: Analizando una variable num√©rica por categor√≠a")
print("Veamos c√≥mo se comporta una m√©trica (likes) seg√∫n la plataforma y la ubicaci√≥n.")

# --- An√°lisis por Plataforma ---
print("\n\n--- AN√ÅLISIS POR PLATAFORMA ---")

# Media de 'likes_count' por plataforma
print("\nüìà Media de 'Likes' por Plataforma:")
media_por_plataforma = df.groupby('platform')['likes_count'].mean().sort_values(ascending=False)
print(media_por_plataforma)

# Moda de 'likes_count' por plataforma (puede haber varias modas)
print("\nüéØ Moda de 'Likes' por Plataforma:")
moda_por_plataforma = df.groupby('platform')['likes_count'].apply(lambda x: x.mode().iloc[0])
print(moda_por_plataforma)

# Distribuci√≥n de 'likes_count' por plataforma (Boxplot)
plt.figure(figsize=(12, 7))
sns.boxplot(x='platform', y='likes_count', data=df, palette='Set2')
plt.title('Distribuci√≥n de Likes por Plataforma')
plt.xlabel('Plataforma')
plt.ylabel('Cantidad de Likes')
plt.show()


# --- An√°lisis por Ubicaci√≥n (Top 5) ---
print("\n\n--- AN√ÅLISIS POR UBICACI√ìN (TOP 5) ---")
top_5_locations = df['location'].value_counts().nlargest(5).index
df_top5_locations = df[df['location'].isin(top_5_locations)]


# Media de 'likes_count' por ubicaci√≥n
print("\nüìà Media de 'Likes' por Ubicaci√≥n (Top 5):")
media_por_ubicacion = df_top5_locations.groupby('location')['likes_count'].mean().sort_values(ascending=False)
print(media_por_ubicacion)

# Moda de 'likes_count' por ubicaci√≥n
print("\nüéØ Moda de 'Likes' por Ubicaci√≥n (Top 5):")
moda_por_ubicacion = df_top5_locations.groupby('location')['likes_count'].apply(lambda x: x.mode().iloc[0])
print(moda_por_ubicacion)


# Distribuci√≥n de 'likes_count' por ubicaci√≥n (Boxplot)
plt.figure(figsize=(15, 8))
sns.boxplot(x='location', y='likes_count', data=df_top5_locations, palette='Set3', order=top_5_locations)
plt.title('Distribuci√≥n de Likes por Ubicaci√≥n (Top 5)')
plt.xlabel('Ubicaci√≥n')
plt.ylabel('Cantidad de Likes')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# PASO 7: AN√ÅLISIS UNIVARIADO - Variables Num√©ricas
# =============================================================================
print("\n\n--- AN√ÅLISIS DE LA DISTRIBUCI√ìN DE VARIABLES NUM√âRICAS ---")

# Seleccionar solo las columnas num√©ricas para los histogramas
variables_numericas = df.select_dtypes(include=np.number).columns.tolist()

# Generar un histograma para cada variable num√©rica
for var in variables_numericas:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[var], kde=True, bins=30)
    plt.title(f'Distribuci√≥n de {var}')
    plt.xlabel(var)
    plt.ylabel('Frecuencia')
    plt.show()
```

```python
# PASO 9: RESUMEN FINAL Y CONCLUSIONES
# =============================================================================
print("\nüéØ PASO 9: Resumen de nuestro an√°lisis")
print("=" * 60)
print("üìù LO QUE HEMOS APRENDIDO:")

print(f"\nüî¢ TAMA√ëO DEL DATASET:")
print(f"   ‚Ä¢ {df.shape[0]} observaciones (publicaciones)")
print(f"   ‚Ä¢ {df.shape[1]} variables")

print(f"\nüìä TIPOS DE VARIABLES:")
print(f"   ‚Ä¢ {len(variables_numericas)} variables cuantitativas: {variables_numericas}")
print(f"   ‚Ä¢ {len(variables_categoricas)} variables cualitativas: {variables_categoricas}")

print(f"\nüìà PRINCIPALES HALLAZGOS:")
for variable in variables_categoricas[:2]:  # Mostrar solo las primeras 2
    categoria_principal = df[variable].value_counts().index[0]
    frecuencia = df[variable].value_counts().iloc[0]
    porcentaje = (frecuencia / len(df)) * 100
    print(f"   ‚Ä¢ {variable}: '{categoria_principal}' es la m√°s com√∫n ({porcentaje:.1f}%)")

if len(variables_numericas) > 0:
    variable_numerica = variables_numericas[0]
    promedio = df[variable_numerica].mean()
    print(f"   ‚Ä¢ {variable_numerica}: promedio de {promedio:.2f}")

print(f"\nüí° PR√ìXIMOS PASOS RECOMENDADOS:")
print("   1. Profundizar en las variables m√°s interesantes")
print("   2. Buscar patrones espec√≠ficos seg√∫n sus preguntas de investigaci√≥n")
print("   3. Considerar an√°lisis m√°s avanzados seg√∫n sus hip√≥tesis")

print("\nüéâ ¬°FELICITACIONES!")
print("Han completado su primer an√°lisis exploratorio de datos")
print("Ahora tienen una comprensi√≥n s√≥lida de su dataset sobre redes sociales")

# Configurar el estilo de los gr√°ficos para que se vean mejor
plt.style.use('default')
sns.set_palette("husl")
```

```python
# PASO 10: MUESTREO ESTRATIFICADO
# =============================================================================
print("\nüîç PASO 10: Creando una muestra representativa (Muestreo Estratificado)")
print("Aseguramos que cada plataforma est√© representada proporcionalmente.")

from sklearn.model_selection import train_test_split

# Supongamos que queremos una muestra del 20% del total de los datos
# 'stratify=df['platform']' asegura que la proporci√≥n de cada plataforma sea la misma en la muestra
df_muestra, _ = train_test_split(df, test_size=0.8, stratify=df['platform'], random_state=42)

print(f"\n‚úÖ Muestra estratificada creada con {df_muestra.shape[0]} observaciones.")

# --- Verificaci√≥n de la estratificaci√≥n ---
print("\n\n--- VERIFICACI√ìN DE LA PROPORCIONALIDAD ---")
print("Veamos si la proporci√≥n de plataformas es similar en el dataset original y en la muestra.")

# Proporci√≥n en el dataset original
print("\nüìä Proporci√≥n en el Dataset Original:")
print(df['platform'].value_counts(normalize=True) * 100)

# Proporci√≥n en la muestra
print("\nüìä Proporci√≥n en la Muestra Estratificada:")
print(df_muestra['platform'].value_counts(normalize=True) * 100)

print("\n\nüëÄ Vistazo a las primeras 5 filas de la muestra:")
display(df_muestra.head())
```

```python
# PASO 13: MUESTREO POR CONGLOMERADOS (CLUSTERS)
# =============================================================================
print("\nüß¨ PASO 13: Creando una muestra por conglomerados")
print("Usaremos la 'location' como nuestro conglomerado.")

# 1. Definir los conglomerados (ubicaciones √∫nicas)
locations = df['location'].unique()

# 2. Seleccionar aleatoriamente algunos conglomerados
# Vamos a seleccionar 5 ubicaciones al azar como nuestros clusters
import random
random.seed(42) # Para reproducibilidad
n_clusters_a_seleccionar = 5
clusters_seleccionados = random.sample(list(locations), n_clusters_a_seleccionar)

print(f"\n‚úÖ Conglomerados seleccionados ({n_clusters_a_seleccionar} ubicaciones): {clusters_seleccionados}")

# 3. Crear la muestra con todos los datos de los conglomerados seleccionados
df_muestra_clusters = df[df['location'].isin(clusters_seleccionados)]

print(f"‚úÖ Muestra por conglomerados creada con {df_muestra_clusters.shape[0]} observaciones.")


# --- Visualizaci√≥n de los resultados ---
print("\n\n--- VISUALIZACI√ìN DEL MUESTREO POR CONGLOMERADOS ---")

# Crear una figura con dos subplots
fig, axes = plt.subplots(1, 2, figsize=(18, 7), sharey=True)
fig.suptitle('Comparaci√≥n de Muestreo por Conglomerados vs. Dataset Original', fontsize=16)

# Gr√°fico para el Dataset Original
sns.countplot(ax=axes[0], y='platform', data=df, order=df['platform'].value_counts().index, palette='viridis')
axes[0].set_title('Dataset Original')
axes[0].set_xlabel('Cantidad de Publicaciones')
axes[0].set_ylabel('Plataforma')

# Gr√°fico para la Muestra por Conglomerados
sns.countplot(ax=axes[1], y='platform', data=df_muestra_clusters, order=df['platform'].value_counts().index, palette='viridis')
axes[1].set_title('Muestra por Conglomerados')
axes[1].set_xlabel('Cantidad de Publicaciones')
axes[1].set_ylabel('')

plt.show()

# Tambi√©n veamos la distribuci√≥n de las ubicaciones en la muestra
plt.figure(figsize=(10, 6))
sns.countplot(y='location', data=df_muestra_clusters, palette='plasma')
plt.title('Distribuci√≥n de Ubicaciones en la Muestra por Conglomerados')
plt.xlabel('Cantidad de Publicaciones')
plt.ylabel('Ubicaci√≥n')
plt.show()
```
